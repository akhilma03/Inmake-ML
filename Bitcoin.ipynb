```python
import numpy as np
import pandas as pd
from  matplotlib import pyplot as plt
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('BitcoinPrice.csv')
```

```python
df.head()
```

```python
df.drop(['Date'],1,inplace=True)
```

```python
df.head()
```

```python
p_day=30
```

```python
df['Predection']=df[['Price']].shift(-p_day)
```

```python
df.head()
```

```python
X=np.array(df.drop(['Predection'],1))
```

```python
X=X[:len(df)-p_day]
```

```python
X.shape
```

```python
X
```

```python
Y=np.array(df['Predection'])
```

```python
Y=Y[:-p_day]
```

```python
Y.shape
```

```python
#split Training
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
P_day_array =np.array(df.drop(['Predection'],1))[-p_day:]
```

```python
P_day_array
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf=RandomForestRegressor(n_estimators=1000,random_state=1)
```

```python
Rf.fit(X_train,y_train)
```

```python
print('Ramdom Forest Accuracy:{:.2f}%'.format(Rf.score(X_test,y_test)*100))
```

```python
#Predection
```

```python
Rf_predict=Rf.predict(X_test)
```

```python
print(Rf_predict)
```
